- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)
```

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
estimators = [('k=6', KMeans(n_clusters=6)),
              ('k=3', KMeans(n_clusters=3)),
              ('k=2', KMeans(n_clusters=2))]
```

```python
fignum = 1
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(10, 7))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    
    ## fit data
    est.fit(X)
    
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2],s=90,
               c=labels.astype(np.float), edgecolor='k')
    ax.view_init(100, 22)
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(name)
    fignum = fignum + 1
```

```python
estimators = [('init=1', KMeans(n_clusters=3,n_init =1,init='random',random_state=131,max_iter=3)),
              ('init=2', KMeans(n_clusters=3,n_init =2,init='random',random_state=131,max_iter=3)),
              ('init=4', KMeans(n_clusters=3,n_init =4,init='random',random_state=131,max_iter=3))]
```

```python
fignum = 1
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(10, 7))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    
    ## fit data
    est.fit(X)
    
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2],s=90,
               c=labels.astype(np.float), edgecolor='k')
    ax.view_init(100, 12)
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(name)
    fignum = fignum + 1
```

```python
# Plot the ground truth
fig = plt.figure(fignum, figsize=(10, 7))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [(iris.target_names[0], 0),
                    (iris.target_names[1], 1),
                    (iris.target_names[2], 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean() + 2,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              size=20,
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
ay = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=ay,s=90, edgecolor='k')
ax.view_init(100, 12)
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Ground Truth')
ax.dist = 12

#fig.show()
```
