# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

# [作業重點]

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype('float64')]
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
numeric_columns = list(filter(lambda x : 'MEDI'not in x,numeric_columns ))
numeric_columns = list(filter(lambda x : 'MODE'not in x,numeric_columns ))
numeric_columns = list(filter(lambda x : 'AVG'not in x,numeric_columns ))
numeric_columns = app_train.loc[:,numeric_columns].count().sort_values(ascending=False)[:40].index.tolist()
print("Numbers of remain {0} columns ".format(len(numeric_columns)))
# 檢視這些欄位的數值範圍
fig, axes = plt.subplots(nrows=6 ,ncols=4,figsize=(13,18))

rowId =0
colId = 0
move=0
for col in numeric_columns:
    """
    Your CODE HERE, make the box plot
    """
    ax = axes[rowId,colId]
    app_train.boxplot(column=[col],ax=ax)
    if colId == 3:
        rowId+=1
    move+=1
    colId = move % 4
    print(col)
plt.show()
```

```python
def ecdf(data):
    n = len(data)
    ind = np.sort(data)
    x = np.arange(1,n+1) / n 
    x = pd.DataFrame(x)
    x.index=ind
    return x
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

## AMT_INCOME_TOTAL

117000000 較有可能是 離群值，有一位客戶的收入遠大於其他人

```python

col ='AMT_INCOME_TOTAL'
df = app_train.loc[~app_train.loc[:,col].isna(),col]
cdf = ecdf(df)
fig,axes =plt.subplots(ncols=2,figsize=(14,5))
axes[0].plot(np.log(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
df.hist(ax=axes[1])
cnts=app_train.loc[:,col].value_counts().sort_index(ascending = False)
plt.show()
plt.scatter(range(len(cnts.iloc[:30].index)),cnts.iloc[:30].index)
plt.show()
list(cnts.index)[:10]
```

## OBS_60_CNT_SOCIAL_CIRCLE

344  為離群值 ，逾期的次數高過其他人

```python
col ='OBS_60_CNT_SOCIAL_CIRCLE'
df = app_train.loc[~app_train.loc[:,col].isna(),col]
cdf = ecdf(df)
fig,axes =plt.subplots(ncols=2,figsize=(14,5))
axes[0].plot(np.log(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
df.hist(ax=axes[1])
cnts=app_train.loc[:,col].value_counts().sort_index(ascending = False)
plt.show()
plt.scatter(range(len(cnts.iloc[:30].index)),cnts.iloc[:30].index)
plt.show()
list(cnts.index)[:10]
```

## AMT_REQ_CREDIT_BUREAU_QRT

261 為離群值，次數超過

```python
col ='AMT_REQ_CREDIT_BUREAU_QRT'
df = app_train.loc[~app_train.loc[:,col].isna(),col]
cdf = ecdf(df)
fig,axes =plt.subplots(ncols=2,figsize=(14,5))
axes[0].plot(np.log(cdf.index+.0001), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
df.hist(ax=axes[1])
cnts=app_train.loc[:,col].value_counts().sort_index(ascending = False)
plt.show()
plt.scatter(range(len(cnts.iloc[:30].index)),cnts.iloc[:30].index)
plt.show()
list(cnts.index)[:10]
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
cdf = ecdf(app_train['AMT_INCOME_TOTAL'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
